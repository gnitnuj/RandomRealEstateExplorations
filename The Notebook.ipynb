```python
#  courtesy of http://www.analyticsvidhya.com/
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=12)
    rolstd = pd.rolling_std(timeseries, window=12)

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
```

#### so the idea behind test_stationarity is to compare p-value to the critical value. (more negative p than critical = more better, excuse the gammer)

```python
test_stationarity(ts_data)
```

```python
# weighted diff

weighted_diff = pd.ewma(ts_data_2_log, halflife=12)
plt.plot(ts_data_2_log)
plt.plot(weighted_diff, color='red')
```

```python
test_stationarity(ts_data_2_log - weighted_diff)
```

#### -2.659725 < -2.573694, so finally I can say that I'm at least 90% confident that the series is stationary

#### decomposing

```python
# credit to http://www.analyticsvidhya.com/
# the below breaks out trend and seasonality into seperate graphs so that it is easier to see their effects

from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts_data_2_log)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# this exact graph is plotted above
plt.subplot(411)
plt.plot(ts_data_2_log, label='Original')
plt.legend(loc='best')

plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)

plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)

plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

```python
# graph the residuals
decompose = residual
decompose.dropna(inplace=True)
test_stationarity(decompose)
```

#### -3.049973 < -2.875207, the best result so far! 95% or more confident that the series is stationary

```python
# try decomposition one more time with (ts_data_2_log - weighted_diff) to see if I can get > 1% confidence

decomposition2 = seasonal_decompose(ts_data_2_log - weighted_diff).resid
decomposition2.dropna(inplace=True)
test_stationarity(decomposition2)
```

#### -3.519472 < -3.461429, so my suspicions were correct!
#### the decomposition2 series is what we should use for modeling as its more that 99% stationary

#### Forecast Model using ARIMA (Auto-Regressive Integrated Moving Averages)

```python
from statsmodels.tsa.stattools import acf, pacf
```

```python
lag_acf = acf(decomposition2, nlags=20)
lag_pacf = pacf(decomposition2, nlags=20, method='ols')
```

```python
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(decomposition2)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(decomposition2)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
```

```python
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(decomposition2)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(decomposition2)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
# okay so the purpose of those graphs are to find out the optimal lag values for generating the ARIMA model.
# looking at the graphs, the optimal values are:  pacf=p=2, acf=q=11
from statsmodels.tsa.arima_model import ARIMA
# http://statsmodels.sourceforge.net/0.6.0/generated/statsmodels.tsa.arima_model.ARIMA.html
```

#### 3 different models based on order forthcoming

The (p,d,q) order of the model for the number of AR parameters, differences, and MA parameters to use.

#### AR Model

```python
# the only thing that'll change is order
model = ARIMA(ts_data_2_log, order=(2, 1, 0))
results_AR = model.fit(disp=-1)  
plt.plot(decomposition2)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-decomposition2)**2))
```

#### MA Model

```python
# the only thing that'll change is order
model = ARIMA(ts_data_2_log, order=(0, 1, 1))  
results_AR = model.fit(disp=-1)  
plt.plot(decomposition2)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-decomposition2)**2))
```

#### Combined Model

```python
model = ARIMA(ts_data_2_log, order=(2, 1, 2))  
results_ARIMA = model.fit(disp=-1)  
plt.plot(decomposition2)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-decomposition2)**2))
```
